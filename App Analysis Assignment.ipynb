# App Analysis Visualization

### Purpose
To visualize and understand user's feelings, attitudes, and sentiments towards the current pregnancy-related, maternal apps targeted to Black and POC women. 

To visualize the users sentiment of the apps, I will complete the following:
1. Collect user review data from the Apple App store
2. Visualize the review content from users into a Word Cloud
3. Perform a sentiment analysis using popular Natural Language Toolkit (NLTK) library
4. Compare the overall user review sentiment scores across apps 
4. Provide design recommendations for culturally-tailored Black maternal mobile apps 

### Tools and Libraries
To scrape data from the Apple App Store app_store_scraper and for sentiment analysis I worked with VADER (Valence Aware Dictionary for Sentiment Reasoning). 

### Analysis Approach
For this analysis, I researched and analyzed 10 maternal mobile applications, that included culturally tailored apps to Black/POC women and social support apps throughout pregnancy to all women: 

Culturally tailored apps:
1. Irth
2. Wolomi
3. What to expect
4. Akina
5. The Journey Pregnancy 

Social Support apps:

6. Peanut
7. Mindful Mamas
8. The Bump
9. Headspace
10. The Baby Centre

### Limitations
While this analysis visualization was a fun exercise to aid in understanding user sentiment and design features that could be used to improve maternal mobile apps, there are areas of limitations to keep in mind: 
- Range of user review data limited/varies with scaper library (server activity/downtime, etc.)  
- Sentiment analysis libraries used were easy to setup. However, though words in sentence were analyzed for sentiment (e.g., negative, positive, etc.), the nuances in the relationship between the words in the sentences lacks interpretation

Thus, this analysis can be viewed as an exercise and could be improved with more robust data and advanced NLT libraries.

### Understanding Rating Skews

### Understand if Sentiment Analysis Scores Align with Ratings

```python
# Initialize VADER sentiment analysis object
analyzer = SentimentIntensityAnalyzer()
```

```python
# Add an Id column for later merging
peanutdf = peanutdf.reset_index().rename(columns={'index': 'Id'})
peanutdf.head()
```

```python
res = {}
for i, row in tqdm(peanutdf.iterrows(), total=len(peanutdf)):
    txt = row['review']
    myid = row['Id']
    res[myid] = analyzer.polarity_scores(txt)
res[0]
```

```python
# Orient the table
peanutdf_sentiment = pd.DataFrame(res).T # T orients the table the correct way
peanutdf_sentiment = peanutdf_sentiment.reset_index().rename(columns={'index': 'Id'})
peanutdf_sentiment.head()
```

```python
# Sentiment scoring merged with original table results
peanutdf_sentiment = peanutdf_sentiment.merge(peanutdf, how='left')
peanutdf_sentiment.head()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
ax_compound = sns.barplot(data=peanutdf_sentiment, x='rating', y='compound')
ax_compound.set_title('Compound Score by Rating')

plt.subplot(1, 3, 2)
ax_pos = sns.barplot(data=peanutdf_sentiment, x='rating', y='pos')
ax_pos.set_title('Positive Score by Rating')

plt.subplot(1, 3, 3)
ax_neg = sns.barplot(data=peanutdf_sentiment, x='rating', y='neg')
ax_neg.set_title('Negative Score by Rating')

plt.tight_layout()
plt.show()
```

### Experimenting with Sentiment Analysis

Using popular NLTK libraries such as TextBlob and VADER, we can conduct a sentiment analysis on the review data.

When working with TextBlob, I found that highly positive reviews that used negative language were identified as 'negative' reviews by TextBlob (see example below). However, VADER seemed to more accurately classify positive ratings that used negative words (i.e. a reviewer discussing black maternal mortalities and how they were glad the app was created).

According to the documentation, Vader segments the scoring into the following categories/thresholds: "positive sentiment: compound score >= 0.05 neutral sentiment: (compound score > -0.05) and (compound score < 0.05) negative sentiment: compound score <= -0.05 ...the compound score is the one most commonly used for sentiment analysis by most researchers"

Since I sometimes wasn't getting accurate identification of positive/negative reviews, I needed to also manually identify reviews myself for my presentation.

## Generating Word Clouds

### Code Generated Word Clouds

```python
all_review_text = " ".join(review for review in peanutdf.review)
negative_reviews_df = peanutdf[peanutdf['rating'] < 3]
negative_review_text = " ".join(review for review in negative_reviews_df.review)
positive_reviews_df = peanutdf[peanutdf['rating'] > 2]
positive_review_text = " ".join(review for review in positive_reviews_df.review)

peanut_stopwords = set(STOPWORDS)
peanut_stopwords.update(["review", "reviews", "app", "peanut", "don"])

peanut_wordcloud_all = WordCloud(stopwords=peanut_stopwords, 
                      width=1600, height=800,
                      min_word_length =3, 
                      background_color="white").generate(all_review_text)

peanut_wordcloud_positive = WordCloud(stopwords=peanut_stopwords, 
                      width=1600, height=800,
                      min_word_length =3, 
                      background_color="white").generate(positive_review_text)

peanut_wordcloud_negative = WordCloud(stopwords=peanut_stopwords, 
                      width=1600, height=800,
                      min_word_length =3, 
                      background_color="white").generate(negative_review_text)

plt.figure( figsize=(20,10) )
plt.imshow(peanut_wordcloud_all, interpolation='bilinear')
plt.axis("off")
plt.show()

plt.figure( figsize=(20,10) )
plt.imshow(peanut_wordcloud_positive, interpolation='bilinear')
plt.axis("off")
plt.show()

plt.figure( figsize=(20,10) )
plt.imshow(peanut_wordcloud_negative, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
irth_review_text = " ".join(review for review in irthdf.review)

irth_stopwords = set(STOPWORDS)
irth_stopwords.update(["review", "reviews", "app", "irth"])

irth_wordcloud = WordCloud(stopwords=irth_stopwords, 
                      width=1600, height=800,
                      min_word_length =3, 
                      background_color="white").generate(irth_review_text)


plt.figure( figsize=(20,10) )
plt.imshow(irth_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
